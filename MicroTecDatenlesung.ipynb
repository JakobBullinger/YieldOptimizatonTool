```python
import pandas as pd

# Function to load and display Excel file
def import_excel(file_path, sheet_name=0):
    """
    Import an Excel file and display its content.

    Parameters:
    file_path (str): Path to the Excel file.
    sheet_name (str or int): Name or index of the sheet to load (default is the first sheet).

    Returns:
    pd.DataFrame: Loaded DataFrame.
    """
    try:
        # Load Excel file
        df = pd.read_excel(file_path, sheet_name=sheet_name)
        print("File loaded successfully!\n")
        print("First 5 rows of the dataset:\n")
        print(df.head())
        return df
    except FileNotFoundError:
        print(f"Error: The file '{file_path}' was not found.")
    except Exception as e:
        print(f"An error occurred: {e}")

# Example usage
if __name__ == "__main__":
    # Correct path to your Excel file using raw string to handle backslashes
    file_path = r"C:\Users\jfxbu\OneDrive - Universitaet St.Gallen\Dokumente\St.Gallen\Praktika\Gelo\Produktivit√§tsbericht\MicroTec\2024_12_5_Formatted.xlsx"
    
    # Call the function to import and display Excel file
    df = import_excel(file_path)
```
